```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("airbnb.csv")
```

```python
print(df.isnull().sum())
```

#handling missing values

```python
df['last review']=pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset=["NAME", "host name"], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

```python
df['price']=df['price'].replace('[\$,]', '', regex = True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

#remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

Visualization

Q1. what is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True, color='#FF5A5F')
plt.title("Distribution of listing price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows a fairly even distribution of listing prices across different price ranges, 
indicating no particular concentration of listing in any specific price range. 
The KDE line helps visulalize this even spread more clearly,
confirming that dataset contains listing with wide variety of prices.

Q2.  How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x="room type", data=df, color="#767676")
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("count")
plt.show()
```

Q3. How are listings distributed across different neighborhoods?

```python
plt.figure(figsize=(12, 8))
sns.countplot(y="neighbourhood group", data=df, color="lightgreen", order=df['neighbourhood group'].value_counts().index)
plt.title('Number Of Listings By Neigbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

Q4. What is the relationship between price and room type?

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='room type', y='price', hue="room type", data=df, palette='Set1')
plt.title('Price V/S Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

Q5. How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number Of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number Of Reviews')
plt.show()
```
